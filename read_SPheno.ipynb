Hermitian term:
    
$\mathcal{L}_{\text{mas}} = \psi^{\dagger} M \psi$
$\to$
$M_{\text{diag}}=U^{-1}MU$

No hermitian term:
 
$\mathcal{L}_{\text{mas}} = \psi_R^{\dagger} M \psi_L = \psi_1^{\dagger} M \psi_2$
$\to$ $\color{red}{M_{\text{diag}}^2=V^{-1}MM^TV = U^{-1}M^TMU}$


ie. the $V$ and $U$ matricess diagonalize the matrices $MM^T$ and $M^TM$ respectively

such that: $\Psi_1^i=V_{ij}\psi_1^j$ and $\Psi_2^i=U_{ij}\psi_2^j$

At one loop level, we obtain neutrino mass matrix:
%
\begin{equation}
\label{eq:righthandedneutrinomassterm}
(\mathcal{M}_{\nu})_{ij}=\frac{1}{(4\pi)^2}\sum_{lk}Y^{\dagger}_{Lil}Y_{Rjl}(V^{lk}_{\Psi_2})^{\dagger}U^{lk}_{\Psi_2}M_k[U^{11}_{\Xi}f(m_1,M_k)U^{21}_{\Xi}+U^{12}_{\Xi}f(m_2,M_k)U^{22}_{\Xi}]
\end{equation}

that can be rewritten as:

\begin{equation}
(\mathcal{M}_{\nu})_{ij}=\sum_ky_{Lil}\Lambda_{l}y_{Rjl}\nonumber\,,
\end{equation}

where:

\begin{equation}
\Lambda_l=\frac{1}{(4\pi)^2}\sum_k(V^{lk}_{\Psi_2})^{\dagger}U^{lk}_{\Psi_2}M_k[U^{11}_{\Xi}f(m_1,M_k)U^{21}_{\Xi}+U^{12}_{\Xi}f(m_2,M_k)U^{22}_{\Xi}]
\end{equation}

\begin{equation}
\label{eq:fms}
f(m_s,M_l)=\frac{m^2_s\ln(m^2_s)-M^2_l\ln(M^2_l)}{m^2_s-M^2_l}\,,\\
\end{equation}

```python
def fk(ms,ml):
    fk = (1./(4.*np.pi)**2)*(ms**2*np.log(ms**2)-ml**2*np.log(ml**2))/(ms**2-ml**2)
    return fk   
```

```python
fk1 = mXi_1*(ZN11*fk(mns_1,mXi_1)*ZN21 + ZN12*fk(mns_2,mXi_1)*ZN22)
fk2 = mXi_2*(ZN11*fk(mns_1,mXi_2)*ZN21 + ZN12*fk(mns_2,mXi_2)*ZN22)
print(fk1,fk2)
```

```python
fk11 = ZR11*ZL11*fk1 + ZR12*ZL12*fk2
fk12 = ZR11*ZL12*fk1 + ZR12*ZL22*fk2
fk21 = ZR12*ZL11*fk1 + ZR22*ZL12*fk2
fk22 = ZR21*ZL21*fk1 + ZR22*ZL22*fk2
#print(fk11,fk12,fk21,fk22)
```

```python
#Element M11 modifications
M11 = YnL11*fk11*YnR11 + YnL21*fk21*YnR11 + YnL11*fk12*YnR21 + YnL21*fk22*YnR21 
M12 = YnL11*fk11*YnR12 + YnL21*fk21*YnR12 + YnL11*fk12*YnR22 + YnL21*fk22*YnR22 
M13 = YnL11*fk11*YnR13 + YnL21*fk21*YnR13 + YnL11*fk12*YnR23 + YnL21*fk22*YnR23 
M21 = YnL12*fk11*YnR11 + YnL22*fk21*YnR11 + YnL12*fk12*YnR21 + YnL22*fk22*YnR21 
M22 = YnL12*fk11*YnR12 + YnL22*fk21*YnR12 + YnL12*fk12*YnR22 + YnL22*fk22*YnR22 
M23 = YnL12*fk11*YnR13 + YnL22*fk21*YnR13 + YnL12*fk12*YnR23 + YnL22*fk22*YnR23 
M31 = YnL13*fk11*YnR11 + YnL23*fk21*YnR11 + YnL13*fk12*YnR21 + YnL23*fk22*YnR21 
M32 = YnL13*fk11*YnR12 + YnL23*fk21*YnR12 + YnL13*fk12*YnR22 + YnL23*fk22*YnR22 
M33 = YnL13*fk11*YnR13 + YnL23*fk21*YnR13 + YnL13*fk12*YnR23 + YnL23*fk22*YnR23 
#print(M11,M12,M13,M21,M22,M23,M31,M32,M33)
```
